```python

import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/0/0c/GoldenGateBridge-001.jpg/1024px-GoldenGateBridge-001.jpg"
image = io.imread(url,0) 
image = cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.figure()

plt.imshow(image)

image.shape
```

```python
mask = np.zeros(image.shape[:2] , np.uint8) 
plt.figure()
plt.imshow(mask , cmap= 'gray')
```

```python
mask[100:300,150:300] = 255
plt.figure()
plt.imshow(mask , cmap= 'gray')
```

```python
mask_image = cv.bitwise_and(image,image,mask=mask)
plt.figure()
plt.imshow(mask_image , cmap= 'gray')
```

```python
masked_image_hist = cv.calcHist([image],channels = [0],mask = mask, histSize=[256] , ranges = [0,256])
plt.figure()
plt.plot(masked_image_hist)
```
