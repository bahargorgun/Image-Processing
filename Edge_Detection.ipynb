```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url = "https://img.freepik.com/premium-photo/big-ben-clock-tower-london-england_79295-6841.jpg?w=2000"
image = io.imread(url,0)
image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
edges = cv.Canny(image = image, threshold1 = 0, threshold2=255)
plt.figure()
plt.imshow(edges, cmap='gray')
plt.axis('off')

median = np.median(image)
print(median)
```

```python
low = int(max(0,(1-0.33)*median))
high = int(min(255,1+0.33)*median)
```

```python
edges_2 = cv.Canny(image = image, threshold1 = low, threshold2=high)
plt.figure()
plt.imshow(edges_2, cmap='gray')
plt.axis('off')
```

```python
#blurred

blurred_image = cv.blur(image,ksize=(5,5))
plt.figure()
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
```

```python
median = np.median(blurred_image)
print(median)
low = int(max(0,(1-0.33)*median))
high = int(min(255,1+0.33)*median)
```

```python
edges_blur = cv.Canny(image = blurred_image, threshold1 = low, threshold2=high)
plt.figure()
plt.imshow(edges_blur, cmap='gray')
plt.axis('off')
```
