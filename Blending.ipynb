```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url =  "https://upload.wikimedia.org/wikipedia/commons/a/a2/View_of_loch_lomond.JPG"
url2 = "https://pix10.agoda.net/hotelImages/149/149052/149052_15012019550024720287.jpg?ca=3&ce=1&s=1024x768"


image = io.imread(url) 
image_2 = io.imread(url2)

print("First picture size: ",image.shape)
print("Second picture size: ",image_2.shape)

n_image = cv.resize(image,(600,600))
n_image_2 = cv.resize(image_2,(600,600))
print("new size for 1 :" , n_image.shape)
print("new size for 2 :",n_image_2.shape)

plt.figure()
plt.imshow(n_image)

plt.figure()
plt.imshow(n_image_2)
```

```python
blended = cv.addWeighted(src1 = n_image , alpha = 0.7, src2= n_image_2 , beta = 0.3,gamma = 0.0)
plt.figure()
plt.imshow(blended)
```
