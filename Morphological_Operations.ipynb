```python

import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url =  "https://a.storyblok.com/f/112937/887x725/8fcc974bbd/friends.png/m/620x0/filters:quality(70)/"
image = io.imread(url,0)
```

```python

plt.figure()
plt.imshow(image)
plt.axis('off'),plt.title('Original')
plt.show()
image.shape
```

```python
kernel = np.ones((5,5), dtype = np.uint8)
result = cv.erode(image , kernel , iterations = 1)

plt.figure()
plt.imshow(result)
plt.axis('off'),plt.title('Erode')
plt.show()
```

```python
result2 = cv.dilate(image,kernel,iterations = 1)

plt.figure()
plt.imshow(result2)
plt.axis('off'),plt.title('Original')
plt.show()
```

```python
#white noise
whiteNoise = np.random.randint(0,2,size = image.shape[:2])
whiteNoise = whiteNoise*255

plt.figure()
plt.imshow(whiteNoise,cmap = 'gray')
plt.axis('off'),plt.title('Noise')
plt.show()
```

```python
noise_Img = whiteNoise + image

plt.figure()
plt.imshow(noise_Img.astype(np.float32))
plt.axis('off'),plt.title('with White Noise')


#Clipping input data to the valid range for imshow with RGB data ([0..1] for floats or [0..255] for integers) warning so i used np.astype()
```

```python
opening = cv.morphologyEx(noise_Img.astype(np.float32),cv.MORPH_OPEN,kernel)

plt.figure()
plt.imshow(opening)
plt.axis('off'),plt.title('Openingl')
```

```python

black_noise = np.random.randint(0,2,size = image.shape[:2])
whiteNoise = whiteNoise*-255 
plt.figure()
plt.imshow(black_noise, cmap = 'gray')
plt.axis('off'),plt.title('Black Noise')
```
