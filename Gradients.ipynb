```python

import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/1/17/Sudoku_puzzle_hard_for_brute_force.jpg"
image = io.imread(url,0) 
plt.figure()
plt.axis('off')
plt.imshow(image, cmap='gray')
```

```python
sobelX = cv.Sobel(image.astype('float32'),ddepth=cv.CV_16S, dx=1 , dy=0 , ksize=5)
plt.figure()
plt.axis('off')
plt.imshow(sobelX, cmap='gray')
```

```python
sobelY = cv.Sobel(image.astype('float32'),ddepth=cv.CV_16S, dx=0 , dy=1 , ksize=5)
plt.figure()
plt.axis('off')
plt.imshow(sobelY, cmap='gray')
```

```python
laplacian = cv.Laplacian(image, ddepth=cv.CV_16S)
plt.figure()
plt.axis('off')
plt.imshow(laplacian, cmap='gray')
```
