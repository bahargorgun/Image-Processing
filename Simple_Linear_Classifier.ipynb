```python
import numpy as np 
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
labels = ["dog", "cat", "panda"]
np.random.seed(1)
```

```python
 W = np.random.randn(3,3072)
 b = np.random.randn(3)
```

```python
url = "https://th.bing.com/th/id/OIP.s6qou2uQK0QlszaSB_yGLAHaGn?pid=ImgDet&rs=1"
image = io.imread(url,0)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_resize = cv.resize(image,(32,32)).flatten()
```

```python
scores = W.dot(image_resize) + b
```

```python
for (label,score) in zip(labels,scores):
  print("[INFO] {}: {:.2F}".format(label,score))

cv.putText(image,"Label: {}".format(labels[np.argmax(scores)]),
(10,30), cv.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0),2)  
```
