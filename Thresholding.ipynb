```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
url = "https://i.guim.co.uk/img/media/2b7ebf35202261a74e13eb6551a452cfcac99c69/193_671_3660_2196/master/3660.jpg?width=1200&height=1200&quality=85&auto=format&fit=crop&s=b68fc2c80cd6fc10c22bbe18683b63ef"
image = io.imread(url) 
image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
#white - 255
```

```python
plt.figure()
plt.imshow(image, cmap="gray")
plt.axis("off")
```

Threshold


```python
_,thresh_img = cv.threshold(image,thresh = 45, maxval = 255, type = cv.THRESH_BINARY)
#60 ve 255 arasn beyaz yapyor

plt.figure()
plt.imshow(thresh_img, cmap = 'gray')
plt.show()
```

Adaptive Thresholding

```python
thresh_img2 = cv.adaptiveThreshold(image,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,4)
plt.figure()
plt.imshow(thresh_img2, cmap = 'gray')
plt.axis('off')
plt.show()
```
