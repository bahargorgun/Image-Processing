```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # for image display
from skimage import io

url =  "https://i.cnnturk.com/i/cnnturk/75/0x555/5438f3adf63099294c6cacb9"
einstein = io.imread(url)


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

face_rect = face_cascade.detectMultiScale(einstein)

for (x,y,w,h) in face_rect:
    cv2.rectangle(einstein,(x,y),(x+w,y+h),(255,255,255),10)
plt.figure(), plt.imshow(einstein,cmap='gray'),plt.axis('off')
```

```python
url2 = 'https://a57.foxnews.com/static.foxnews.com/foxnews.com/content/uploads/2022/10/640/320/barcelona.jpg?ve=1&tl=1'
url3 = 'https://media-cldnry.s-nbcnews.com/image/upload/newscms/2017_25/2044296/170619-fc-barcelona-team-ed-355p.jpg'
barcelona = io.imread(url3)
face_rect = face_cascade.detectMultiScale(barcelona,minNeighbors=7)

for (x,y,w,h) in face_rect:
    cv2.rectangle(barcelona,(x,y),(x+w,y+h),(255,255,255),7)
    
plt.figure(figsize=(15,15)), plt.imshow(barcelona,cmap='gray'),plt.axis('off')
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()


    if ret :
        face_rect = face_cascade.detectMultiScale(frame)
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,255),10)
        frame = cv2.flip(frame, 1)
    
        cv2.imshow('face detect',frame)

    if cv2.waitKey(1) & 0xFF == ord('q'): break  
cap.release()
cap.destroyAllWindows()
```
